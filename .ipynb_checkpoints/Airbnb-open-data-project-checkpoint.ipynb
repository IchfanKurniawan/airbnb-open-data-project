### <span id='1.3'>1.3 Data Cleaning  

```python
# From the "To do"s above we could clean unwanted entries:

    # drop negative values 'minimum_nights', 'availability_365'
    # outliers in 'minimum_nights' 'availability_365'
    # drop low variance columns 'country', 'country_code', 'license'
    # drop duplication
```

```python
neg_index = df[(df['minimum_nights'] < 0) | (df['availability_365'] < 0)].index
df.drop(labels=neg_index, axis=0, inplace=True)
```

```python
outlrs_cols = ['minimum_nights', 'availability_365']

for col in outlrs_cols:
    interquartile = df[col].quantile(0.75) - df[col].quantile(0.25)
    lower_bound = df[col].quantile(0.25) - 1.5*interquartile
    higher_bound = df[col].quantile(0.75) + 1.5*interquartile
    
    outlrs_index = df[(df[col] < lower_bound) | (df[col] > higher_bound)].index
    df.drop(labels=outlrs_index, axis=0, inplace=True)
```

```python
# From the "To do"s above we could preprocess & derive new features:

# Derive New Features:
    # is_review
    # binning features: 'number_of_reviews' 'reviews_per_month' 'calculated_host_listings_count'
    # merge perfect correlated features # perfect correlation between price & service_fee into total_fee price_over_service_fee


# Preprocess:
    # fill the missing values in each feature if possible
    # group together little portion categorical values
    # the NaN value could be impute with 'None' or group together with others or left them to be NaN
    # transform skewed right: 'minimum_nights' 'availability_365'
    # scaling
```

```python
# fill missing values

df['name'].fillna('No Name', inplace=True)
df['host_identity_verified'].fillna('None', inplace=True)
df['host_name'].fillna('None', inplace=True)
df['neighbourhood'].fillna('Other', inplace=True)
df['neighbourhood_group'].fillna('Other', inplace=True)
df['instant_bookable'].fillna('None', inplace=True)
df['cancellation_policy'].fillna('None', inplace=True)
df['construction_year'].fillna(df['construction_year'].median(), inplace=True)
df['minimum_nights'].fillna(df['minimum_nights'].median(), inplace=True)
df['number_of_reviews'].fillna(0, inplace=True)
df['reviews_per_month'].fillna(0, inplace=True)
df['review_rate_number'].fillna(0, inplace=True)
df['calculated_host_listings_count'].fillna(0, inplace=True)
df['availability_365'].fillna(df['availability_365'].median(), inplace=True)
df['house_rules'].fillna('None', inplace=True)
```

```python
# fill missing value with knnimputer for price & service fee

num_col = df.select_dtypes(include='number')
cat_col = df.select_dtypes(exclude='number')

from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=5, weights="uniform")
num_col = imputer.fit_transform(num_col)
```

```python
num_col = pd.DataFrame(num_col, 
                       columns=df.select_dtypes(include='number').columns)
df_prep = pd.concat([num_col, cat_col], axis=1)
```

```python
# derive new features

df_prep['is_review'] = np.vectorize(lambda x: 0 if x==0 else 1)(df_prep['number_of_reviews'])
df_prep['bin_number_of_reviews'] = pd.qcut(df_prep['number_of_reviews'], 3, labels=False)
df_prep['bin_reviews_per_month'] = pd.qcut(df_prep['reviews_per_month'], 3, labels=False)
df_prep['bin_calculated_host_listings_count'] = pd.qcut(df_prep['calculated_host_listings_count'].rank(method='first'), 3, labels=False)
df_prep['total_fee'] = df_prep['price'] + df_prep['service_fee']
df_prep['price_over_service_fee'] = np.vectorize(lambda x, y: np.round(x/y,3))(df_prep['price'], df_prep['service_fee'])
```

```python
# for simpler model, group together categorical values

df_prep['neighbourhood_group'] = df_prep['neighbourhood_group'].replace(to_replace=['manhatan', 'brookln', 'Staten Island', 'Bronx'] ,
                                       value=['Other','Other','Other','Other'])
```

```python
# transform these column into near normal distribution

df_prep['log_minimum_nights'] = np.log10(df_prep['minimum_nights'] + 1)
df_prep['log_availability_365'] = np.log10(df_prep['availability_365'] + 1)
```

```python
# this is the target column

df_prep['bin_total_fee'] = pd.qcut(df_prep['total_fee'], 3, labels=[0,1,2])
```
